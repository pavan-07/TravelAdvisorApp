```python
ContentData = df[['business_id', 'name', 'categories', 'attributes', 'stars', 'review_count']].copy()
ContentData.fillna({'categories': '', 'attributes': 'Unknown'}, inplace=True)

```

```python
ContentData['content'] = ContentData['name'] + ' ' + ContentData['categories'] + ' ' + ContentData['attributes']
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(ContentData['content'])
```

```python
CosineSimilarity = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_content_recommendations(name, CosineSimilarity=CosineSimilarity, ContentData=ContentData, threshold=0.1):

    # get index of the restaurent
    idx = ContentData.index[ContentData['name'] == name].tolist()
    if not idx:
        print(f"There were no restaurants present as '{name}'")
        return []

    idx = idx[0]
    SimilarityScores = list(enumerate(CosineSimilarity[idx]))
    
    # filter out items with similarity below the threshold
    SimilarityScores = [(i, score) for i, score in SimilarityScores if score > threshold]
    SimilarityScores = sorted(SimilarityScores, key=lambda x: x[1], reverse=True)
    
    # get top 10 similar restaurents
    TopSimilarRestaurents = SimilarityScores[1:11]
    
    # Get the indices and names of the top similar restaurents
    SimilarIndices = [i[0] for i in TopSimilarRestaurents]
    SimilarRestaurents = ContentData.iloc[SimilarIndices][['name', 'review_count', 'stars']]
    return SimilarRestaurents
```

```python
RestaurantName = "Domino's Pizza"
ContentRecommendations = get_content_recommendations(RestaurantName, threshold=0.5)

print(f"Listed below is the Content-Based Recommendations for {RestaurantName}:")
print(ContentRecommendations)
```
